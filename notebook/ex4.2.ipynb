# Scatter plot

```python
# set seaborn style
sns.set()
# Scatter plot
# byYear = birth_rate_yrly.groupby('year')
byYearMean = birth_rate_yrly.groupby('year').mean()
# When groupby operation is performed, by default the column on which groupby is done becomes the index.
byYearMean.reset_index(inplace=True)
# byYear.count()
# Mean birth rate for each year
# byYearMean

plt.figure(figsize=(15, 9))
plt.scatter(byYearMean['year'], byYearMean['rate'])
plt.title("Scatter plot showing the mean birth rate by Year", fontsize=18)
plt.ylabel("Mean Annual Birth Rate", fontsize=15)
plt.xlabel("Year", fontsize=15)
```

# Bubble Chart

```python
# Bubble chart

plt.figure(figsize=(25, 9))
sns.scatterplot(data=crime_rate, x="state", y="murder", size="population",alpha=0.5, sizes=(200,800))
plt.title("Bubble chart showing the murder rate by States in 2005", fontsize=18)
plt.ylabel("Murder Rate", fontsize=15)
plt.xlabel("States", fontsize=15)
plt.xticks(rotation=45)

plt.show()
```
