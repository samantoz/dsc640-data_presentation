```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
output_width = 1000
#output_width = 80 #//*** Normal Output width
pd.set_option("display.width", output_width)
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
# Load the source file

# currDir = os.getcwd()
# print(f"Current Folder :{currDir}")

world_population_df = pd.read_excel("world-population.xlsm",header=0)
world_population_df
```

```python
# Draw plot


plt.figure(figsize=(15,10))
plt.xlabel("Year", fontsize=20)
plt.ylabel("Population(in billions)", fontsize=24)
plt.title("World Population growth \n 1960 to 2010", fontsize=30)
plt.plot(world_population_df["Year"], world_population_df["Population"], color='skyblue', alpha=1)

plt.show()

plt.figure(figsize=(15, 10))
plt.xlabel("Year", fontsize=20)
plt.ylabel("Population(in billions)", fontsize=24)
plt.title("World Population growth \n 1960 to 2010", fontsize=30)

# Subsetting by decades only to show the steps clearly
df_world_pop_by_decade = world_population_df.loc[world_population_df['Year'].isin([1960,1970,1980,1990,2009])]

df_world_pop_by_decade

X = df_world_pop_by_decade["Year"]
Y = df_world_pop_by_decade["Population"]

plt.step(x=X,y=Y, color='skyblue', alpha=1)

plt.show()
```
